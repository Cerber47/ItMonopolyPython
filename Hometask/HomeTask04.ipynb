## Упражнение 4.1

Требуется написать функцию deposit_info(money, time_period), принимающую два аргумента: money - количество денег, которые хочет вложить пользователь, и time_period - на какой срок. Функция должна возвращать количество денег, которые будет иметь пользователь по прошествии указанного периода по следующей схеме:

- Если срок меньше шести месяцев, то его выигрыш составит 5% в месяц от первоначальной суммы
- Если срок больше шести месяцев, то его выигрыш линейно растет по формуле 0.5 * m + 5, где m - продолжительность вклада. Его выигрыш не может быть больше 40% в месяц от первоначальной суммы
- Если сумма свыше 500 000 рублей, то пользователь дополнительно получает 15 000 рублей за каждые 500 000 положенных рублей

money - целое число в рублях, time_period - время в месяцах.

## Упражнение 4.2

Требуется написать функцию triangle_length(), вычисляющую периметр треугольника. В качестве аргументов она должна принимать либо три числа - длины трех сторон треугольника, либо длины двух сторон. В том случае, когда пользователь указывает два аргумента, функция должна считать, что треугольник прямоугольный, а два аргумента - длины двух катетов.

## Упражнение 4.3

Требуется написать функцию cesar_cipher(n, s), где n - сдвиг, s - строчка на английском языке. функция должна возвращать строчку, зашифрованную шифром Цезаря. Функция должна работать и при отрицательных значения n

## Упражнение 4.4

Требуется написать функцию two_letter_generation(s), где s - строчка. Функция должна брать от функции и возвращать генератор. От строчки нужно брать две буквы и возвращать их без пробелов и знаков препинания в нижним регистре

## Упражнение 4.5
