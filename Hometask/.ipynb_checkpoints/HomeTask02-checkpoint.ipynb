# Задача 1 - Дехитомия

Метод деления отрезка пополам - самый простой способ численного решения уравнения. Если f(x) - заданная функция, нуль которой требуется найти на отрезке a и b, то если f(a) и f(b) разных знаков то между a и b существует хотя бы один нуль f(x). Далее выбирается точка с - посередине между a и b. Нуль функции f(x) может находится таким образом либо на отрезке [a-с] либо на отрезке [c-b] либо в точке с. Таким образом можно делать много раз, до тех пор пока мы не приблизимся к решению на определенную зараняя epsilone (когда изучаемый отрезок станет меньше epsilone), либо найдем точку в которой f(x) равна нулю, либо достигнем максимума итераций.

Требуется написать программу, которая принимает на вход числа a и b (a<b), epsilone и выводит на экран 0 функции f(x). Функцию f(x) требуется задать самостоятельно внутри кода. ITERATION_MAXIMUM также задать самостоятельно. Функцию для примера можно использовать такую $ x^2 - 4.001$

```python
def f(x):
    return x**2 - 4.001
```
