## Упражнения 5.1 

Требуется написать программу, которая будет бесконечно проводить арифметические операции, до тех пор как пользователь не напишет 0. Все операции - бинарны. Если пользователь сделал неверную просьбу - не паниковать, и вежливо сказать ему об этом и дать попробовать еще раз. Требуемые операции - Сложение, деление, умножение и разность двух чисел

## Упражнение 5.2

Требуется написать программу, которая принимает на вход название файла, а затем строчку, которую программа запишет в этот файл, после чего программа завершит свою работу.

## Упражнение 5.3

Требуется написать программу, считывающая все введенные из командной строки числа, и выводит те из них, которые делятся на 2 или на 5 на экран
