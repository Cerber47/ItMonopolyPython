## Упражнения 5.1 

Требуется написать программу, которая будет проводить арифметические операции до тех пор, пока пользователь не напишет 0. Все операции - бинарны. Если пользователь сделал неверную просьбу, - не паниковать, вежливо сказать ему об этом и дать попробовать еще раз. Требуемые операции: сложение, деление, умножение и разность двух чисел

## Упражнение 5.2

Требуется написать программу, которая принимает на вход название файла, а затем строчку, которую программа записывает в этот файл, после чего завершает свою работу.

## Упражнение 5.3

Требуется написать программу, которая считывает все введенные из командной строки числа и выводит на экран те из них, что делятся на 2 или на 5
